### Split our documents so we don't run into token issues.

```python
text_splitter = RecursiveCharacterTextSplitter(separators=["\n\n", "\n"], chunk_size=2000, chunk_overlap=250)
texts = text_splitter.create_documents(dataset['transcript'])
```

```python
# Your api key should be an environment variable, or else put it here
# We are using a chat model in case you wanted to use gpt4
llm = ChatOpenAI(temperature=0)
```

### Prompt for extract data

```python
summary_output_options = {
    'number' : """
     - Number format
     - Only one output
     """,

    'blood_pressure' : """
    - Blood pressure format
    - Example: 120/80
    """,

     'biner' : """
    - write: 有 for yes, 無 for no, and 不明 for unclear
    """,

    'one_sentence' : """
     - Only one sentence
     - In Japanese
    """,
    
    'bullet_points': """
     - Bullet point format
     - Separate each bullet point with a new line
     - Each bullet point should be concise
     - In Japanese
    """,
    
    'short' : """
     - A few short sentences
     - Do not go longer than 4-5 sentences
     - In Japanese
    """,
    
    'long' : """
     - A verbose summary
     - You may do a few paragraphs to describe the transcript if needed
     - In Japanese
    """
}
```

```python
'''
You are an assistant who helps a nurse record the condition of an elderly person in Japanese from "input_documents".
The goal is to find information about "blood pressure" from "input_documents".
If there are no words or sentences specifying blood pressure information, interpret the "input_documents" analysis to read "Systolic pressure (the pressure when the heart contracts) / Diastolic pressure (the pressure when the heart contracts)" /Dystolic blood pressure (pressure when the heart contracts)" Determine blood pressure information in the format. beat)"
'''
```

```python
template="""

あなたは、看護師が「input_documents」から高齢者の状態を日本語で記録するのを手伝うアシスタントです。
目標は、「input_documents」 から「血圧」に関する情報を見つけることです。
血圧情報を指定する単語や文章がない場合は、「input_documents」分析を解釈して、「収縮期血圧（心臓が収縮するときの圧力）/拡張期血圧（心臓が収縮しているときの圧力）/拡張期血圧（心臓が収縮しているときの圧力）」の形式で血圧情報を決定してください。ビート）」

「input_documents」 以外の内容で返信しないでください。 分からない場合は「～」と言ってください			

Respond with the following format
{output_format}

"""

ai_temperature_prompt = AIMessagePromptTemplate.from_template(template)

human_template="{text}" # Simply just pass the text as a human message
human_message_prompt_combine = HumanMessagePromptTemplate.from_template(human_template)

chat_prompt_combine = ChatPromptTemplate.from_messages(messages=[ai_temperature_prompt, human_message_prompt_combine])
```

```python
chain = load_summarize_chain(llm,
                             chain_type="map_reduce",
                             combine_prompt = chat_prompt_combine,
                             verbose=True
                            )
```

```python
user_selection = 'number'

output = chain.run({
                    "input_documents": texts,
                    "output_format" : summary_output_options[user_selection]
                   })

print(output)
```
